```python
import pandas as pd
import numpy as np
import openpyxl
import requests

import pyEX as p

# Import iex cloud api token
from config import iex_api_token
```

```python
# Connect to IEX cloud api endpoint
c = p.Client(api_token=iex_api_token, version='stable')

ticker = 'AAPL'

df = c.dividendsDF(symbol=ticker)

df
```

```python
nly_input = c.dividendsDF(symbol='NLY')
df = pd.concat([df, nly_input], ignore_index=True)

df
```

```python
# Create a list of stocks that you know issue stock dividends
div_list = ['VFC', 'DVN', 'BBY', 'VZ', 'T', 'INTC', 'PM', 'MMM']

for i in div_list:
    div_data = c.dividendsDF(symbol=i)
    df = pd.concat([df,div_data], ignore_index=True)

df.head(5)
```

```python
# Drop the extra columns 
df = df.drop(columns = ['adrFee', 'coupon', 'declaredCurrencyCD', 'declaredGrossAmount', 'id', 'key', 'figi', 
'fiscalYearEndDate', 'secondExDate', 'secondPaymentDate', 'securityType', 'toDate', 'toFactor', 'date', 'updated', 'parValue',
'parValueCurrency', 'periodEndDate', 'recordDate', 'refid', 'registrationDate', 'taxRate', 'subkey',
'isApproximate', 'isCapitalGains', 'isDAP', 'isNetInvestmentIncome', 'marker', 'netAmount', 'notes', 'optionalElectionDate',
'description', 'flag', 'fromFactor', 'fxDate', 'installmentPayDate', 'announceDate', 'created', 'grossAmount',
'declaredDate', 'lastUpdated', 'countryCode'])

# Set column order
column_order = ['symbol', 'exDate', 'paymentDate', 'frequency', 'currency', 'amount']
df = df[column_order]

df
```
