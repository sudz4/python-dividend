When you’re looking for ways to receive regular dividend payments, you generally have two options: stocks that pay dividends and funds that hold stocks that pay dividends.

```python
# Connect to IEX cloud api endpoint
c = p.Client(api_token=iex_api_token, version='stable')

# Test api connection
ticker = 'AAPL' # IMPORTANT - not all stocks issue dividends
timeframe = '6m'
test_df = c.dividendsDF(symbol=ticker)

test_df
```

```python
# set the columns and covert from index object to list
col = test_df.columns
col_list = list(col)

# Create a pandas data frame with cols 
div_df = pd.DataFrame(columns=col_list)

div_df
```

```python
# Create a list of stocks that you know issue stock dividends
div_list = ['VFC', 'DVN', 'BBY', 'NLY', 'VZ', 'T', 'INTC', 'PM', 'WBA', 'MMM']

for i in div_list:
    data_input_df = c.dividendsDF(symbol=i)
    div_df = div_df.append(data_input_df, ignore_index=True)
```

```python
# Look at the staging df
div_df
```

```python
# Drop columns 
div_df = div_df.drop(columns = ['adrFee', 'coupon', 'declaredCurrencyCD', 'declaredGrossAmount', 'id', 'key', 'figi', 
'fiscalYearEndDate', 'secondExDate', 'secondPaymentDate', 'securityType', 'toDate', 'toFactor', 'date', 'updated', 'parValue',
'parValueCurrency', 'periodEndDate', 'recordDate', 'refid', 'registrationDate', 'taxRate', 'subkey',
'isApproximate', 'isCapitalGains', 'isDAP', 'isNetInvestmentIncome', 'marker', 'netAmount', 'notes', 'optionalElectionDate',
'description', 'flag', 'fromFactor', 'fxDate', 'installmentPayDate', 'announceDate', 'created', 'grossAmount',
'declaredDate', 'lastUpdated', 'countryCode'])

div_df
```

```python
# set column order
column_order = ['symbol', 'exDate', 'paymentDate', 'frequency', 'currency', 'amount']
div_df = div_df[column_order]

div_df
```
